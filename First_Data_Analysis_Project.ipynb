```python
#import necessary packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset using pandas
df = pd.read_csv("StudentScores.csv")
```

```python
#now use describe method to describe the data set it gives us a lot of information about the data
df.describe()
```

```python
#tells us data type of the data of the column
df.info()
```

```python
#now find how many values are null
df.isnull().sum()
```

```python
#fill the null values if necessary but here no score value is null so no need to fill
#then drop the unnamed column 
df = df.drop("Unnamed: 0", axis = 1)
```

```python
df.head()
```

```python
#now lets do some real data analysis
```

Gender Distribution:

```python
#countplot
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Number Of Male And Female Students")
plt.show()
```

#from the above chart we have analysed that the number of female is more then that of males

```python
# now lets create a graph representing impact of parents education in students studies
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore": 'mean'})
```

```python
gb
#here we groupped the parent edu with mean of scores and result is like this
```

```python
#now lets plat in heatmap
sns.heatmap(gb, annot = True) #if we dont use annot = true then the number or marks will not be shown
plt.title("Relationship Between Parents Education And Students Scores")
plt.show()
```

```python
#now lets check maretial status impact o scores
# same as above we will create a groupby 
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore": 'mean'})
gb1
```

```python
sns.heatmap(gb1, annot = True) #if we dont use annot = true then the number or marks will not be shown
plt.title("Relationship Between Parents Maritial Status And Student's Score")
plt.show()
```

```python
# we concluded that there is almost no impact or neglegible impact
```
